## Problem 1: Mobile Recharge Validation System
### Problem Statement
Create a function to validate a mobile recharge request.
#### Rules:
- Recharge amount must be ≥ ₹50
- Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid
#### Real-Time Use
- Telecom recharge apps
- Payment gateways

#### Hint
- Use function for validation
- Use while loop for retry

```python
def recharge(amount):
    plans = [199,299,399,599]
    min_amount = 50

    while True:
        try:
            amount = int(input('Enter recharge amount'))

            if amount < min_amount:
                print(f'Invalid! Amount must be at least {min_amount}. Try again.')
                continue

            if amount in plans:
                print(f'Recharge successful! Plan {amount} activated.')
                break

            else:
                print("Invalid plan! Available plans: 199, 299, 399, 599. Try again.")

        except ValueError:
            print('Please enter a valid number.')

recharge('amount')
```

## Problem 2: Inventory Reorder Alert System
### Problem Statement
Given a dictionary of products and stock quantities:
- Print Reorder Alert if stock < 15
- Print Stock OK otherwise

#### Real-Time Use
- Warehouse systems
- Retail management

#### Hint
- Use function
- Use for loop on dictionary

```python
def stock(inventory):
    for p, q in inventory.items():
        if q < 15:
            print(f'{p}: Reorder alert!')
        else:
            print(f'{p}: Stock Ok')

inventory_data = {'Coffee': 10, 'Milk': 50, 'Books': 30, 'Orea': 14, 'Chips': 26}
stock(inventory_data)
```

## Problem 3: Student Result Processing System
### Problem Statement
- Create a function that:
- Accepts a list of student marks
- Calculates average marks
- Prints Pass if average ≥ 50 else Fail

#### Real-Time Use
- Online exam systems
- Evaluation portals

#### Hint
- Use loop to calculate total
- Return result from function

```python
def student(marks):
    avg_marks = sum(marks)/len(marks)
    if avg_marks >= 50:
        print(f'Average: {avg_marks} -> Pass!')
    else:
        print(f'Average: {avg_marks} -> Fail, try hard next time. Good luck!!')

stu1 = [50,40,80,77,94]
stu2 = [75,84,64,83,58]
stu3 = [70,40,60,41,35]
students = stu1, stu2, stu3
for stu in students:
    student(stu)
```

## Problem 4: Cab Fare Estimator with Retry Option
### Problem Statement
Create a cab fare calculator:
- Base fare = ₹50
- ₹12 per km
- Peak hour → 25% extra
- Allow retry using while loop

#### Real-Time Use
- Ride-hailing apps
- Transport systems

#### Hint
- Function for fare calculation
- While loop for retry

```python
def fare_estimate(distance, peak_hour=False):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    
    if peak_hour:
        fare *= 1.25
    return fare

while True:
    try:
        km = float(input('Enter distance travelled (in km)'))
        peak = input("Is it peak hour? (yes/no)").strip().lower()

        if peak not in ["yes", "no"]:
            print("Invalid input for peak hour. Please enter 'yes' or 'no'")
            continue

        total_fare = fare_estimate(km, peak == "yes")
        print(f"Estimated Cab Fare: Rs.{total_fare:.2f}")
        break   # Exit loop after successful calculation
    
    except ValueError:
        print("Invalid input! Please enter a numeric value for distance.")
```

## Problem 5: Employee Attendance Eligibility Checker
### Problem Statement
Create a function that:
- Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if ≥ 75% else Not Eligible

#### Real-Time Use
- HR payroll systems
- Internship tracking

#### Hint
- Use loop to count present days
- Use return statement

```python
def attendance_eligibility(attend_list):
    total_days = len(attend_list)
    present_days = 0

    for day in attend_list:
        if day == "P":
            present_days +=1

    percentage = (present_days/total_days)*100

    if percentage >= 75:
        return f"Attendance percentage: {percentage:.2f} -> Eligible"
    else:
        return f"Attendance percentage: {percentage:.2f} -> Not Eligible"

emp1 = ["P", "P", "P", "P", "P", "A", "A", "P", "P", "P", "P", "P", "A", "P", "P"]
emp2 = ["A", "A", "P", "A", "P", "P", "P", "P", "P", "P", "P", "A", "P", "A", "P"]

emps = emp1, emp2
for emp in emps:
    print(attendance_eligibility(emp))
```

## Problem 6: Password Strength Checker
### Problem Statement
Create a function to check password strength.
#### Rules:
- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@#$)

#### Real-Time Use
- Login systems
- Security validation

#### Hint
- Use loop over string
- Use conditions inside function

```python
def password_strength():
    while True:
        password = input("Enter a password: ")

        # min length
        if len(password) < 8:
            print("Password must contain at least 8 characters. Try again.")
            continue
        
        # at least 1 digit
        if not any(ch.isdigit() for ch in password):
            print("Password must contain at least 1 digit. Try again.")
            continue

        # at least 1 spl character
        if not any(ch in "@#$" for ch in password):
            print("Password must contain at least 1 special character (@,#,$). Try again.")
            continue

        print("Strong password accepted!!")
        break

# run the function
password_strength()
```
